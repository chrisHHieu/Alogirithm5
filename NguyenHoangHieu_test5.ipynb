```python
import numpy as np
import pandas as pd
```

```python
# Tính Delta P
def calculate_delta_p(p_start):
    return (1.5 - 10 * p_start) / 45

# Hàm tính xác suất không thắng P_all_lose
def calculate_p_all_lose(p_start, delta_p, num_rounds=10):
    # Tạo dãy giá trị P và xác suất thua cho từng lần chơi
    p_values = np.array([p_start + i * delta_p for i in range(num_rounds)])
    lose_probs = 1 - p_values
    p_all_lose = np.prod(lose_probs)
    return p_all_lose, p_values, lose_probs
```

```python
# Danh sách các giá trị P_start để kiểm tra từ 0.01 đến 1 theo bước nhảy là 0.01
p_start_list = np.arange(0.01, 1, 0.01)
results = []
```

```python
for p_start in p_start_list:
    delta_p = calculate_delta_p(p_start)
    p_all_lose, p_values, lose_probs = calculate_p_all_lose(p_start, delta_p)
    
    # Nếu xác suất không thắng P_all_lose <= 0.05, lưu kết quả
    if p_all_lose <= 0.05:
        result = {
            "P_start": round(p_start, 2),
            "Delta_P": round(delta_p, 4),
            "P_all_lose": round(p_all_lose, 4),
            "Thỏa điều kiện": "Có"  # Thêm cột "Condition Met"
        }
        for i, p in enumerate(p_values):
            result[f'P_value_{i+1}'] = round(p, 4)
        
        results.append(result)

df_results = pd.DataFrame(results)
```
